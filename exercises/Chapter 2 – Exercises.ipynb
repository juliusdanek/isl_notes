```python
# our new dataframe
df = pd.read_csv('../datasets/auto.csv')
# Are there any missing values? (https://stackoverflow.com/questions/29530232/python-pandas-check-if-any-value-is-nan-in-dataframe)
df.isnull().values.any()
```

```python
# let's take a look at the rows
df.iloc[:10]
```

```python
df.describe()
```

```python
# horsepower doesn't seem to be stored as an int, let's check
df.dtypes
```

```python
# correct! It is stored as an object. Let's try to convert it. 
try:
    df['horsepower'] = df['horsepower'].astype('int')
except Exception as e:
    print repr(e)
```

```python
# apparently some of the values in horsepower are missing. Let's drop the rows with missing data.
# You can find the rows with ? like so: df.loc[df['horsepower'] = '?']
df = df[df['horsepower'] != '?']
# let's convert horsepower to an int
df['horsepower'] = df['horsepower'].astype('int')
df.describe()
```

```python
df.hist(figsize=(12,12))
plt.show()
```

a) Let's take a look at the general data. Judging from the histograms and the describe function, we can reasonably deduct that we should take a closer look at origin and cylinders.

```python
df['origin'].value_counts()
```

```python
df['cylinders'].value_counts()
```

We can assume that origin is a qualitative variable expressed in numbers, same with cylinders. 

```python
# b) + c)
# let's just use 'describe' on the quantitative columns to find the range, mean and std
df[df.columns.difference(['origin', 'cylinders'])].describe()
```

```python
# d)
# excluding observations 10 to 85 and describing
df[~df.index.isin(range(10,86))].describe()
```

```python
# e)
pd.tools.plotting.scatter_matrix(df, figsize=(15,15))
plt.show()
```

One can reasonably deduce a few relationship, here I will highlight three:

1. Weight & mpg â€“ negatively correlated, higher mpg = lower weight
2. Displacement & mpg â€“ negatively correlated, higher displacement = lower weight
3. Displacement & weight â€“ positively correlated, higher weight = higher displacement

### f)

Based on the scatterplots from e), we can assume that all other variables can be useful in predicting mpg. All of them seem to have some effect / relationship with mpg. Some relationships might be weaker than others (year, acceleration, origin)

```python
# our new dataframe. Source: https://forge.scilab.org/index.php/p/rdataset/source/tree/master/csv/MASS/Boston.csv
df = pd.read_csv('../datasets/boston.csv')
df.describe()
```

There are 506 rows in this dataset. The dataset contains the housing values in the suburbs of Boston. You can find a description of the dataset in [this pdf](https://cran.r-project.org/web/packages/MASS/MASS.pdf) on page 20.

```python
# let's do some scatterplots
pd.tools.plotting.scatter_matrix(df, figsize=(15,15), alpha=0.2)
plt.show()
```

b)

There are some relationships clearly visible:

1. Number of Rooms and Median Value
2. Nitrogen Oxide concentration (nox) and age
3. Distance to employment centers (dis) and age

```python
# c)
# There seems to be a weak relationship between distance and crime, age and crime
df.plot(kind='scatter', x='crim', y='dis', color='r')    
df.plot(kind='scatter', x='crim', y='age', color='g')
df.plot(kind='scatter', x='crim', y='rad', color='b')
df.plot(kind='scatter', x='crim', y='tax')
df.plot(kind='scatter', x='crim', y='ptratio')
plt.show()
```

However, these observations might not be statistically significant and none of them look like they have a strong relationship.
